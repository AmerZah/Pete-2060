```python
#get rid of outliers for log of the data           
q1, q3= np.percentile(df01,[25,75])
iqr = q3 - q1 
df01_outliers = df01[~((df01 <(q1 - 1.5 * iqr ))| (df01 > (q3 + 1.5 * iqr ))).any(axis=1)]
df01_outliers
```

```python
#get outliers for SGoil for log of the data
q1, q3= np.percentile(df01.SGoil,[25,75])
iqr = q3 - q1 
lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)
upper_bound, lower_bound
```

```python
#get outliers for SGgas for log of the data
q1, q3= np.percentile(df01.SGgas,[25,75])
iqr = q3 - q1 
lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)
upper_bound, lower_bound
```
